```python
# importing data
df = pd.read_csv('diabetic_data.csv')
```

```python
# This cell deals with medications. Medications are sorted by class type.
# If any medications within that class are increased, the med class will recieve a 2.
# If any are decreased, they receive a -2.
# Otherwise, if medication perscriptions remain steady, the medication class receives a 1.
# If no medications in that class are perscribed, they receive a 0.

biguanides = []
meglitinides = []
sulfonylureas = []
thiazolidinediones = []
alpha_glucosidase_inhibitors = []

for value in df['metformin']:
    # if any are increased or decreased, that takes precedence over entire category
    # there should not be two instances of Down and Up in the same medication area
    if value == 'Down' or value in df['glyburide-metformin'] == 'Down' or value in df['glipizide-metformin'] == 'Down' or value in df['metformin-rosiglitazone'] == 'Down' or value in df['metformin-pioglitazone'] == 'Down':
        biguanides.append(-2)
    elif value == 'Up' or value in df['glyburide-metformin'] == 'Up' or value in df['glipizide-metformin'] == 'Up' or value in df['metformin-rosiglitazone'] == 'Up' or value in df['metformin-pioglitazone'] == 'Up':
        biguanides.append(2)
    elif value == 'Steady' or value in df['glyburide-metformin'] == 'Steady' or value in df['glipizide-metformin'] == 'Steady' or value in df['metformin-rosiglitazone'] == 'Steady' or value in df['metformin-pioglitazone'] == 'Steady':
        biguanides.append(1)
    else:
        biguanides.append(0)

for value in df['repaglinide']:
    # if any are increased or decreased, that takes precedence over entire category
    # there should not be two instances of Down and Up in the same medication area
    if value == 'Down' or value in df['nateglinide'] == 'Down':
        meglitinides.append(-2)
    elif value == 'Up' or value in df['nateglinide'] == 'Up':
        meglitinides.append(2)
    elif value == 'Steady' or value in df['nateglinide'] == 'Steady':
        meglitinides.append(1)
    else:
        meglitinides.append(0)

for value in df['chlorpropamide']:
    # if any are increased or decreased, that takes precedence over entire category
    # there should not be two instances of Down and Up in the same medication area
    if value == 'Down' or value in df['glimepiride'] == 'Down' or value in df['acetohexamide'] == 'Down' or value in df['glipizide'] == 'Down' or value in df['glyburide'] == 'Down' or value in df['tolbutamide'] == 'Down' or value in df['tolazamide'] == 'Down' or value in df['glipizide-metformin'] == 'Down' or value in df['glyburide-metformin'] == 'Down' or value in df['glimepiride-pioglitazone'] == 'Down':
        sulfonylureas.append(-2)
    elif value == 'Up' or value in df['glimepiride'] == 'Up' or value in df['acetohexamide'] == 'Up' or value in df['glipizide'] == 'Up' or value in df['glyburide'] == 'Up' or value in df['tolbutamide'] == 'Up' or value in df['tolazamide'] == 'Up' or value in df['glipizide-metformin'] == 'Up' or value in df['glyburide-metformin'] == 'Up' or value in df['glimepiride-pioglitazone'] == 'Up':
        sulfonylureas.append(2)
    elif value == 'Steady' or value in df['glimepiride'] == 'Steady' or value in df['acetohexamide'] == 'Steady' or value in df['glipizide'] == 'Steady' or value in df['glyburide'] == 'Steady' or value in df['tolbutamide'] == 'Steady' or value in df['tolazamide'] == 'Steady' or value in df['glipizide-metformin'] == 'Steady' or value in df['glyburide-metformin'] == 'Steady' or value in df['glimepiride-pioglitazone'] == 'Steady':
        sulfonylureas.append(1)
    else:
        sulfonylureas.append(0)

for value in df['pioglitazone']:
    # if any are increased or decreased, that takes precedence over entire category
    # there should not be two instances of Down and Up in the same medication area
    if value == 'Down' or value in df['rosiglitazone'] == 'Down' or value in df['troglitazone'] == 'Down' or value in df['metformin-rosiglitazone'] == 'Down' or value in df['metformin-pioglitazone'] == 'Down':
        thiazolidinediones.append(-2)
    elif value == 'Up' or value in df['rosiglitazone'] == 'Up' or value in df['troglitazone'] == 'Up' or value in df['metformin-rosiglitazone'] == 'Up' or value in df['metformin-pioglitazone'] == 'Up':
        thiazolidinediones.append(2)
    elif value == 'Steady' or value in df['rosiglitazone'] == 'Steady' or value in df['troglitazone'] == 'Steady' or value in df['metformin-rosiglitazone'] == 'Steady' or value in df['metformin-pioglitazone'] == 'Steady':
        thiazolidinediones.append(1)
    else:
        thiazolidinediones.append(0)

for value in df['acarbose']:
    # if any are increased or decreased, that takes precedence over entire category
    # there should not be two instances of Down and Up in the same medication area
    if value == 'Down' or value in df['miglitol'] == 'Down':
        alpha_glucosidase_inhibitors.append(-2)
    elif value == 'Up' or value in df['miglitol'] == 'Up':
        alpha_glucosidase_inhibitors.append(2)
    elif value == 'Steady' or value in df['miglitol'] == 'Steady':
        alpha_glucosidase_inhibitors.append(1)
    else:
        alpha_glucosidase_inhibitors.append(0)

df['insulin'].replace({'Down': -2, 'Up': 2, 'Steady': 1, 'No': 0,}, inplace=True)


df = df.drop(['encounter_id', 'patient_nbr', 'gender', 'weight', 'admission_source_id', 'payer_code', 'diag_3', 'metformin', 'repaglinide',	'citoglipton', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone', 'tolazamide', 'examide', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone',	'metformin-pioglitazone', 'diabetesMed'], axis=1)
#admission source ID dropped because admission type ID gives same info in more simplified terms

df['biguanides'] = biguanides
df['meglitinides'] = meglitinides
df['sulfonylureas'] = sulfonylureas
df['thiazolidinediones'] = thiazolidinediones
df['alpha_glucosidase_inhibitors'] = alpha_glucosidase_inhibitors

# Used to confirm order of Down/Up does not affect change in final matrix
# biguanides2 = []
# for value in df['metformin']:
#     # if any are increased or decreased, that takes precedence over entire category
#     # there should not be two instances of Down and Up in the same medication area
#     if value == 'Up' or value in df['glyburide-metformin'] == 'Up' or value in df['glipizide-metformin'] == 'Up' or value in df['metformin-rosiglitazone'] == 'Up' or value in df['metformin-pioglitazone'] == 'Up':
#         biguanides2.append(2)
#     elif value == 'Down' or value in df['glyburide-metformin'] == 'Down' or value in df['glipizide-metformin'] == 'Down' or value in df['metformin-rosiglitazone'] == 'Down' or value in df['metformin-pioglitazone'] == 'Down':
#         biguanides2.append(-2)
#     elif value == 'Steady' or value in df['glyburide-metformin'] == 'Steady' or value in df['glipizide-metformin'] == 'Steady' or value in df['metformin-rosiglitazone'] == 'Steady' or value in df['metformin-pioglitazone'] == 'Steady':
#         biguanides2.append(1)
#     else:
#         biguanides2.append(0)

# set1 = set(biguanides)
# set2 = set(biguanides2)
# diff = list(set1.symmetric_difference(set2))
# print(diff)

df.to_csv('modified_diabetic_data.csv', index=False)
```
